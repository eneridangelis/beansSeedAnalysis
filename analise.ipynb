```python
# Getting anormal seeds

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Get the current working directory (where the code file is located)
current_directory = os.getcwd()+'/bad'

# Define the naming pattern for the CSV files:
file_pattern = 'b_{}.csv' 

# Determine the range of file numbers or a list of specific file numbers you want to load:
start_file_number = 1
end_file_number = 426  

# Initialize an empty list to store the matrices
matrix_bad = pd.DataFrame({})

# Loop through the file numbers and load each CSV file
for i in range(start_file_number, end_file_number + 1):
    file_name = file_pattern.format(i)
    file_path = os.path.join(current_directory, file_name) 
    
    # Load the CSV file into a DataFrame and exclude the first column
    df = pd.read_csv(file_path,sep=";", header=None)  
    matrix = df[df.columns[1:]]  
    matrix['ID'] = 'b_'+str(i)
    matrix_bad = pd.concat([matrix_bad, matrix], axis = 0)
```

```python
# Getting normal seeds

# Get the current working directory (where the code file is located)
current_directory = os.getcwd()+'/good'

# Define the naming pattern for the CSV files:
file_pattern = 'g_{}.csv'  

# Determine the range of file numbers or a list of specific file numbers you want to load:
start_file_number = 1
end_file_number = 470  

# Initialize an empty list to store the matrices
matrix_good = pd.DataFrame({})

# Loop through the file numbers and load each CSV file
for i in range(start_file_number, end_file_number + 1):
    file_name = file_pattern.format(i)
    file_path = os.path.join(current_directory, file_name)  
    
    # Load the CSV file into a DataFrame and exclude the first column
    df = pd.read_csv(file_path,sep=";", header=None)  
    matrix = df[df.columns[1:]]  
    matrix['ID'] = 'b_'+str(i)
    matrix_good = pd.concat([matrix_good, matrix], axis = 0)
```

```python
df_matrix_bad = matrix_bad.iloc[:, :-1]
df_matrix_good = matrix_good.iloc[:, :-1]

mean_bad = df_matrix_bad.mean()
mean_good = df_matrix_good.mean()
```

```python
import seaborn as sns

# Create a new DataFrame that combines the max values from both DataFrames
combined_max_values = pd.concat([mean_bad, mean_good], axis=1)
combined_max_values.columns = ['Anormal Seeds', 'Normal Seeds']

# Create a boxplot for the combined max values
plt.figure(figsize=(12, 6))
sns.boxplot(data=combined_max_values)
plt.title('Boxplot of Mean Values')

# Show the plot
plt.show()

```

```python
max_bad = df_matrix_bad.max()
max_good = df_matrix_good.max()

# Create a new DataFrame that combines the max values from both DataFrames
combined_max_values = pd.concat([max_bad, max_good], axis=1)
combined_max_values.columns = ['Anormal Seeds', 'Normal Seeds']

# Create a boxplot for the combined max values
plt.figure(figsize=(12, 6))
sns.boxplot(data=combined_max_values)
plt.title('Boxplot of Max Values')

# Show the plot
plt.show()
```

```python
min_bad = df_matrix_bad.min()
min_good = df_matrix_good.min()

# Create a new DataFrame that combines the max values from both DataFrames
combined_max_values = pd.concat([min_bad, min_good], axis=1)
combined_max_values.columns = ['Anormal Seeds', 'Normal Seeds']

# Create a boxplot for the combined max values
plt.figure(figsize=(12, 6))
sns.boxplot(data=combined_max_values)
plt.title('Boxplot of Min Values')

# Show the plot
plt.show()

```

```python
# Extract the identifier column (last column)
identifier_column = matrix_bad.iloc[:, -1]

# Remove the identifier column from the DataFrame for calculating the mean
df_without_identifier = matrix_bad.iloc[:, :-1]

# Group the data by the identifier column and calculate the mean for each group
grouped = df_without_identifier.groupby(identifier_column).mean()

# Transpose the grouped DataFrame for plotting
# grouped = grouped.T

plt.imshow(grouped, cmap='viridis', aspect='auto', origin='lower')

# Adicionar uma barra de cores para interpretar os valores
plt.colorbar()

# Adicionar rótulos aos eixos, se desejar
plt.title('Mean Values by Seed per Collumn - Anormal Seeds')
plt.xlabel('Colunas')
plt.ylabel('Sementes')

# Mostrar o gráfico
plt.show()
```
